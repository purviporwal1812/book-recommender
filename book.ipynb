```python
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
```

```python
path = 'https://drive.google.com/uc?id=1htr2hz7BFXPK17YRfARLirplmCYDDH2B&export=download'
books = pd.read_csv(path)
```

```python
ratings = pd.read_csv('https://drive.google.com/uc?id=1woPthBj9LwcQZwYVdl5PGZYPNDjimWQu&export=download')
```

```python
users = pd.read_csv('https://drive.google.com/uc?id=1-UbsRrx-HTXJhq3IQP51WPJ5A503UglG&export=download')
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating' : 'num_ratings' }, inplace=True)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
popular_df
```

```python
popular_df = popular_df.merge(books , on='Book-Title', suffixes=('_left', '_right')).drop_duplicates('Book-Title')
```

```python
ratings_with_name.columns
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
imp_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(imp_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python

pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    if book_name in pt.index:
        index = np.where(pt.index==book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

        data = []
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Title'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

            data.append(item)

        return data
    else:
        print(f"Book '{book_name}' not found in the dataset.")
```

```python
recommend('Welcome to the World, Baby Girl!')
```
